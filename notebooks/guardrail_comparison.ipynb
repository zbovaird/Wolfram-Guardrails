Compare English LLM policy guardrails against Wolfram symbolic rules over the same prompt dataset.

````python
from pathlib import Path
import json
import sys

REPO_ROOT = Path('.').resolve()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from eval.comparison_runner import run_comparison

dataset = REPO_ROOT / 'datasets' / 'evaluation_prompts.jsonl'
output_dir = REPO_ROOT / 'results'

summary = run_comparison(dataset, output_dir, use_mock=True, use_mirror_only=True)
print(json.dumps(summary, indent=2))
````

````python
# Inspect disagreements
disagreements_path = output_dir / 'disagreements.jsonl'
if disagreements_path.exists():
    for line in disagreements_path.read_text(encoding='utf-8').splitlines():
        if line.strip():
            print(json.dumps(json.loads(line), indent=2))
            print('---')
````
